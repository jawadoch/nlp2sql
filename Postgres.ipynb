```python
from sqlalchemy import create_engine

try:
    engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
    connection = engine.connect()
    print("Connection successful!")
except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd

# Step 1: Read the file with a tolerant encoding
try:
    df = pd.read_csv('online_retail.csv', encoding='latin1')
except Exception as e:
    print(f"Error reading the file: {e}")

# Step 2: Save the DataFrame back to a CSV file with UTF-8 encoding
try:
    df.to_csv('online_retail1.csv', index=False, encoding='utf-8')
    print("File saved successfully with UTF-8 encoding.")
except Exception as e:
    print(f"Error saving the file: {e}")
```
